### 4.2 Data Wrangling
**Data Loading**<br>

```python
# monthly total recreational visits
visits = pd.read_csv('data/monthly_visitation.csv')
# monthly regular gasoline prices
gas_price = pd.read_csv('data/GASREGW.csv')
# monthly total population
pop = pd.read_csv('data/POP.csv')
# monthly unemployment rate
unemploy_rate = pd.read_csv('data/UNRATE.csv')
# monthly personal saving rate
saving_rate = pd.read_csv('data/PSAVERT.csv')

visits.head()
```

**Data Merging**<br>
To combine all datasets into a single dataframe, all data is merged to 'visits' by the date column and duplicate date column is dropped.

```python
# merge gasoline prices to total recreational visits by date
visits = visits.merge(gas_price, left_on = "date", right_on ="DATE")
# rename GASREGW to gas_price
visits.rename(columns={"GASREGW":"gas_price"}, inplace = True) 
# drop excessive date columns 
visits.drop(columns = ["DATE"], inplace = True)

# merge total population to total recreational visits by date
visits = visits.merge(pop, left_on = "date", right_on ="DATE")
#rename POP to pop
visits.rename(columns={"POP":"pop"}, inplace = True) 
# drop excessive date columns 
visits.drop(columns = ["DATE"], inplace = True)

# merge unemployment rate to total recreational visits by date
visits = visits.merge(unemploy_rate, left_on = "date", right_on ="DATE")
 #rename UNRATE to unemploy_rate
visits.rename(columns={"UNRATE":"unemploy_rate"}, inplace = True)
# drop excessive date columns 
visits.drop(columns = ["DATE"], inplace = True)

# merge personal saving rate to total recreational visits by date
visits = visits.merge(saving_rate, left_on = "date", right_on ="DATE")
 #rename PSAVERT to saving_rate
visits.rename(columns={"PSAVERT":"saving_rate"}, inplace = True)
# drop excessive date columns 
visits.drop(columns = ["DATE"], inplace = True)

# print a few rows of this dataset for checking
visits
```

**Data Manipulation** <br>
To study the seasonal effect on the visitation demand, the categorical variables for month will be converted to dummy variables using `One-Hot Encoding`.

```python
#use One-Hot Encoding to get dummeries from month (mnth)
visits_new = pd.get_dummies(visits, columns = ['mnth'])
# check the new dataFrame
visits_new.info()
```

The requirement for K-1 dummy variables in regression is achieved by treating December as the reference level and dropping it from dataframe.

```python
# Use December as reference level and drop it from the dataframe 
visits_final = visits_new.drop(['mnth_DEC'], axis=1)
# double check the result
visits_final.info()
```

## 5. Methodology
***
With the insight from data exploration, this study adopts tree-based machine learning algorithms: CART, Random Forest and XGBoost, as these methods have no assumption on data distribution and correlation between variables. The conceptual demand model of national park visitation is: <br>
$$Monthly Visitation = f(Travel Cost,Population,Seasonal Dummies,Recession)$$ <br>
Models will be evaluated by R-square and RMSE values to estimate the statistical significance, then feature importance will be analysed along with partial dependence and individual conditional expectation plot.

### 5.1 Splitting Into Train and Test Subsets
The default 75:25 setting of `train_test_split` is applied with random split to avoid selection bias by introducing randomness.

```python
# set random state as 100
random_state_split = 100
# Splitting data into training and testing data
train_x, test_x, train_y, test_y = train_test_split(visits_final.drop(['cnt'], axis = 1), visits_final.cnt, random_state=random_state_split)
```

Specification of training and testing is checked by:

```python
# check rows and columns of output 
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
# check the index of train_x and train_y 
print(train_x.index.identical(train_y.index))
print(test_x.index.identical(test_y.index))
```

### 5.2 Classification and Regression Tree (CART)
With no assumption about the data relationship and requirement of transformation, CART is suitable for the the large range and sparese values' distribution of this dataset. <br>
<br>
**Hyperparameters Tuning**<br>
Cross validation determine the optimal settings and provide a comprehensive use of data for model training with consideration of average performance, which is applicable for sparsely distributed data. First, the default setting of tree depth is derived from `get_depth`.

```python
# CART: using default settings
cart_default = DecisionTreeRegressor(random_state=0) #random_state = 0: same train and test sets across different executions
cart_default.fit(train_x, train_y)
# print the tree depth
print("Tree depth: {}".format(cart_default.get_depth()))
```

Having the default tree depth as 21, the range of max_depth is set aruond the value and default setting is used for min_samples_split. 

```python
# values of max_depth (set the range around 21) and min_samples_split (default = 2)
hyperparameters = {'max_depth':[10,15,20,25,30], 'min_samples_split':[2,4,6,8,10]}
# set random state to ensure the reproducibility
randomState_dt = 10000
dt = DecisionTreeRegressor(random_state=randomState_dt)

# cv=5 by default, which means 5-fold cross-validation
clf = GridSearchCV(dt, hyperparameters)
clf.fit(train_x, train_y)

# query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (clf.best_params_)
print ("The best score is: ")
print (clf.best_score_)
```

**Final Model**<br>

```python
# Training the final CART with derived parameters from cross validation
dt_final = DecisionTreeRegressor(max_depth=clf.best_params_['max_depth'], min_samples_split=clf.best_params_['min_samples_split'], random_state=randomState_dt)
dt_final.fit(train_x, train_y)
```

```python
# Check the performance of model 
print("R2 on the training data:")
print(dt_final.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(dt_final.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, dt_final.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, dt_final.predict(test_x), squared=False))
print("Normalised RMSE on the training data:")
print(mean_squared_error(train_y, dt_final.predict(train_x), squared=False)/np.mean(train_y))
print("Normalised RMSE on the testing data:")
print(mean_squared_error(test_y, dt_final.predict(test_x), squared=False)/np.mean(test_y))
```

**Feature Importance**<br>

```python
# estimate the importances function 
imp_cart = rfpimp.importances(dt_final, test_x, test_y)
print(imp_cart)
# barplot of the feature importance
viz = rfpimp.plot_importances(imp_cart)
viz.view()
```

### 5.3 Random Forest (RF)
Random forest ables to compute more compact and accurate result by predicting from a combination of CART. The same workflow from CART will be applied to both RF and XGBoost.<br>
<br>
**Hyperparameters Tuning**<br>

```python
# values of max_depth and min_samples_split
hyperparameters = {'max_depth':[10,15,20,25,30], 'min_samples_split':[2,4,6,8,10]}
# set random state to ensure the reproducibility
randomState_dt = 10000
rf = RandomForestRegressor(random_state=randomState_dt)

# cv=5 by default, which means 5-fold cross-validation
clf = GridSearchCV(rf, hyperparameters)

clf.fit(train_x, train_y)

# query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (clf.best_params_)
print ("The best score is: ")
print (clf.best_score_)
```

**Final Model**<br>

```python
# Training the final random forest
rf_final = RandomForestRegressor(max_depth=clf.best_params_['max_depth'], min_samples_split=clf.best_params_['min_samples_split'], random_state=randomState_dt)
rf_final.fit(train_x, train_y)
```

```python
# Check the performance of model 
print("R2 on the training data:")
print(rf_final.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(rf_final.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, rf_final.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, rf_final.predict(test_x), squared=False))
print("Normalised RMSE on the training data:")
print(mean_squared_error(train_y, rf_final.predict(train_x), squared=False)/np.mean(train_y))
print("Normalised RMSE on the testing data:")
print(mean_squared_error(test_y, rf_final.predict(test_x), squared=False)/np.mean(test_y))
```

**Feature Importance**<br>

```python
# estimate the importances function 
imp_rf = rfpimp.importances(rf_final, test_x, test_y)
print(imp_rf)
# barplot of the feature importance
viz = rfpimp.plot_importances(imp_rf)
viz.view()
```

### 5.4 XGBoost
XGBoost perform gradient boosting based on the residual from previous CART which inaccurate prediction receives higher weight and importance.<br>
<br>
**Hyperparameters Tuning**<br>

```python
# values of max_depth and n_estimators
hyperparameters = {'max_depth':[10,15,20,25,30], 'n_estimators':[50,100,150,200,250]}
# set random state to ensure the reproducibility
randomState_xgb = 10000
xgb = XGBRegressor(random_state=randomState_xgb)

# cv=5 by default, which means 5-fold cross-validation
gscv_xgb = GridSearchCV(xgb, hyperparameters)

gscv_xgb.fit(train_x, train_y)

# we can query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (gscv_xgb.best_params_)
print ("The best score is: ")
print (gscv_xgb.best_score_)
```

**Final Model**<br>

```python
# Training the final XGBoost
xgb_final = XGBRegressor(max_depth=gscv_xgb.best_params_['max_depth'], n_estimators=gscv_xgb.best_params_['n_estimators'], random_state=randomState_xgb)
xgb_final.fit(train_x, train_y)
```

```python
# Check the performance of model 
print("R2 on the training data:")
print(xgb_final.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(xgb_final.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, xgb_final.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, xgb_final.predict(test_x), squared=False))
print("Normalised RMSE on the training data:")
print(mean_squared_error(train_y, xgb_final.predict(train_x), squared=False)/np.mean(train_y))
print("Normalised RMSE on the testing data:")
print(mean_squared_error(test_y, xgb_final.predict(test_x), squared=False)/np.mean(test_y))
```

**Feature Importance**<br>

```python
# estimate the importances function
imp_xg = rfpimp.importances(xgb_final, test_x, test_y) # permutation
print(imp_xg)
# barplot of the feature importance
viz = rfpimp.plot_importances(imp_xg)
viz.view()
```

### 6.2 Discussion
**Limitation** <br>
As this study perform analysis with simple demand model, a limitation is the number of independent variables considered is not sufficient to derive a comprehensive conclusion, such as insufficient demogrpaahic and park-specific variables. Moreover, prediction derived from this study only generally considers the monthly visitation without separating visitation number by activity type and trip duration. <br>

Another limitation is the adopted time frame is only 27 years with consideration of completeness and reliability of data. This may restrain the forecasting ability of model due to the size of datasets and variations that learned by the model, which requires further integration with longer time frame. <br>

With the development of management partnership strategies, there is a need to understand the heritage values of individual parks for conservation (Goetcheus et al., 2018).
